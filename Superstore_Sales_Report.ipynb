#**Superstore Sales Analysis** (PowerBI)

With the Superstore Sales Analysis Dashboard in Power BI, businesses can gain a comprehensive understanding of their sales performance, identify areas for improvement, optimize resource allocation, and make data-driven decisions to drive growth and profitability.

Below listed are some of the important objectives/features that we will consider for this analysis:

1) Sales Performance Overview

2) Regional Analysis

3) Product Analysis

4) Customer Segmentation


---



## Brief description of the dataset

The dataset contains around 51k rows and 13 columns

The dataset contains the following **13** columns:

*   Order ID: contains the unique order IDs
*   Date: date of sale
*   Customer Name
*   Segment: 3 unique categories --> consumer, corporate, home office
*   Country: country of sales
*   Market: continent/market of sales
*   Product ID: unique product IDs
*   Category: categories of different products (3 unique: furniture, office supplies, technology)
*   Sub-Category: sub-categories of the product
*   Product Name: product name
*   Sales: sales of product
*   Quantity: quantity of product sale
*   Profit: profit made (contains both positive and negative values)

### Customer Segmentation

The "Segment" column in the dataset represents the segmentation of customers into different categories based on their characteristics or buying behavior. It helps to classify customers into distinct groups to better understand their preferences, needs, and purchasing patterns. In this specific dataset, the "Segment" column has three unique categories:

Consumer: This segment typically includes individual customers who make purchases for personal use.

Corporate: This segment comprises customers who purchase products on behalf of a company or organization.

Home Office: This segment represents customers who have a home-based office or business and make purchases related to their work or business needs.
